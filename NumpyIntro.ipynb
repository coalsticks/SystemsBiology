```python
def find_atg(seq):
    pos = seq.find("ATG")
    # account for potential errors
    assert pos >= 0, "there is no ATG in the sequence"
    return pos
```

```python
find_atg(seq)
```

```python
def find_stop(seq):
    for pos in range(0, len(seq), 3):
        triplet = seq[pos:pos+3]
        if codontab[triplet] == "*":
            return pos
```
